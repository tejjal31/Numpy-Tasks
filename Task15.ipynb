Part E - Reshaping & Data Structure

Task E1: Reshape the electricity data into a single row representing weekly usage ( Shape: 1x28).
Task E2: Reshape the same data into 2 rows and 14 columns. Ensure no data is lost or duplicated.

````python
import numpy as np

# ================================
# Step 1: Create electricity array
# ================================

# Generate 4x7 random electricity usage data
electricity = np.random.randint(120, 651, (4, 7))

print("Electricity Usage:")
print(electricity)



# ================================
# Part E: Reshaping & Data Structure
# ================================

# Task E1: Reshape into 1 Ã— 28 (single row)
reshape_1x28 = electricity.reshape(1, 28)

print("Reshaped to 1x28:")
print(reshape_1x28)


# Task E2: Reshape into 2 Ã— 14
reshape_2x14 = electricity.reshape(2, 14)

print("\nReshaped to 2x14:")
print(reshape_2x14)
````
