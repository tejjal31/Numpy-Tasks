Task D1: Thinking task
Compare each hostel's daily usage with its own weekly average. Create a boolean matrix where:
- True: Above hostel average
- False: Below or equal to hostel average

Requirement - requires broadcasting of a (4,) array across a (4,7) array. 

````python
# ================================
# Task D1: Compare with hostel averages
# ================================
import numpy as np

# ================================
# Step 1: Create electricity array
# ================================

# Generate 4x7 random electricity usage data
electricity = np.random.randint(120, 651, (4, 7))

print("Electricity Usage:")
print(electricity)



# Step 1: Calculate weekly average per hostel (row-wise)
hostel_avg = np.mean(electricity, axis=1)

# Step 2: Reshape averages to (4,1) for broadcasting
hostel_avg = hostel_avg.reshape(4, 1)

# Step 3: Compare each value with its hostel average
comparison = electricity > hostel_avg

# Display results
print("Hostel Weekly Averages:")
print(hostel_avg)

print("\nBoolean Comparison Matrix:")
print(comparison)
````

# ðŸ§  Task D1 â€” Comparing Usage with Hostel Averages

## ðŸŽ¯ Goal

For each hostel, compare daily electricity usage with its **own weekly average**.

We create a boolean matrix:

* **True** â†’ above average
* **False** â†’ below or equal to average

---

## ðŸ” Step 1: Calculate Hostel Averages

```python
hostel_avg = np.mean(electricity, axis=1)
```

* `np.mean()` calculates averages
* `axis=1` â†’ row-wise (per hostel)
* Result shape = **(4,)**

---

## ðŸ” Step 2: Reshape for Broadcasting

```python
hostel_avg = hostel_avg.reshape(4, 1)
```

We convert shape:

```
(4,) â†’ (4,1)
```

This allows NumPy to broadcast averages across 7 days.

---

## ðŸ” Step 3: Element-wise Comparison

```python
comparison = electricity > hostel_avg
```

NumPy broadcasts the (4,1) array across the (4,7) array.

Each hostel is compared with its own average.

---

## â­ Broadcasting Idea

```
(4,1) â†’ stretched to â†’ (4,7)
```

Each row average is copied across its row.

---

## ðŸŽ¯ Result

We get a **boolean matrix** showing which days are above average.

---

## ðŸŽ¯ Viva Line

> Broadcasting allows a smaller array to expand across a larger array for element-wise comparison.
