## 1️⃣ Load Dataset (Local CSV)


```python
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

# UPDATE THIS PATH if needed
DATA_PATH = "AmesHousing.csv"

df = pd.read_csv(DATA_PATH)
df.shape
```

## 3️⃣ Identify Target Variable (Regression)

From dataset documentation:
- Target variable = `SalePrice`

This is a **continuous numeric variable**, so:

➡️ This is a **regression problem** (for now).

```python
TARGET_COL = "SalePrice"

df[TARGET_COL].describe()
```

## 4️⃣ Separate Feature Types (First Pass)

We classify features **by data type**, not meaning (yet).

This is a mechanical step.

```python
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df.select_dtypes(exclude=[np.number]).columns.tolist()

len(numeric_features), len(categorical_features)
```

```python
numeric_features[:10], categorical_features[:10]
```

## 5️⃣ Feature Taxonomy (Meaning-Based Classification)

Now we classify features by **semantic meaning**, not dtype.

Why?
Because preprocessing decisions depend on *what a feature represents*,

not just how it is stored.


```python
# Manual grouping (will refine later)
identifier_features = ["Id"] if "Id" in df.columns else []

target_feature = [TARGET_COL]

numeric_continuous = [
    col for col in numeric_features 
    if col not in identifier_features + target_feature
]

numeric_continuous[:10]
```

## 6️⃣ Missing Value Analysis (NO fixing yet)

We only observe:
- which features have missing values
- how severe the missingness is

```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df) * 100).round(2)

missing_table = (
    pd.DataFrame({
        "missing_count": missing_count,
        "missing_percent": missing_percent
    })
    .sort_values(by="missing_percent", ascending=False)
)

missing_table[missing_table["missing_count"] > 0].head(15)
```

## 7️⃣ Initial Observations (Write Before Code Changes)

At this stage, we should already notice:
- Some features are mostly missing
- Some are rarely missing
- Numeric and categorical features behave very differently
