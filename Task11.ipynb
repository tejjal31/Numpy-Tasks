## âœ… Part 1: Understanding `np.random.randint()`

`np.random.randint()` generates random integers.

### Syntax

```python

```

Where:

* **low** â†’ starting value (included)
* **high** â†’ ending value (excluded)
* **size** â†’ shape of the output array

## ðŸŽ¯ Viva Line

> `np.random.randint` generates random integers within a specified range and arranges them into a matrix of a given shape.


````python
import numpy as np

# -----------------------------
# Task A1: Create electricity array
# -----------------------------

# Create a 4x7 array with random electricity usage
# Values range from 120 to 650
electricity = np.random.randint(120, 651, (4, 7))

# Print the generated array
print("Electricity Usage Array:")
print(electricity)

# -----------------------------
# Task A2: Print array attributes
# -----------------------------

# Print shape of the array
print("\nShape:", electricity.shape)

# Print number of dimensions
print("Number of dimensions:", electricity.ndim)

# Print total number of elements
print("Total elements:", electricity.size)

# Print datatype of elements
print("Data type:", electricity.dtype)
````
