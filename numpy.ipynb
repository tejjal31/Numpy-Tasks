```python
#arr[start : stop : step]
print(arr[1:3]) #from index 1 to 3 (3 not included)
print(arr[:2]) #from start to 2 (2 not included)
print(arr[::2]) #start to end ,step=2 
```

COPY()

```python
a = np.array([1,2,3])
b = a.copy()   #copy
# a.copy()-> creatyes a new , seperate array in memory
# b is independent of a 
b[0] = 99
print(a)
print(b)
```

```python
a = np.array([1,2,3])
b = a       # this logic is not safe
# here ,it is not a copy , just a another reference
#a and b are linked here
b[0] =99
print(a)
```

BASIC ARRAY OPERATIONS (VECTOR POWER)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

SCALAR OPERATIONS

```python
#scalar is a single number.. apply operation with single number to every element of an array
print(a + 10)
print(a * 2)
```

unfus UNIVERSAL FUNCTIONS

```python
arr = np.array([1,4,9,16]) #creates a numpy array
print(np.sqrt(arr)) # square root of each element
print(np.log(arr)) # natuiral log (ln) of each element
print(np.exp(arr)) # e^x (Exponential) of each element
print(np.max(arr)) #maximum value in array
print(np.min(arr)) #minimum value in array
print(np.mean(arr)) #average (mean) of array values
print(np.sum(arr)) #sum of all elements
print(np.std(arr)) #standard deviation 
```

**MATRIX MULTIPLICATION**

```python
# matrix multiplication using np.dot() by taking dot product
A = np.array([[1,2],
             [3,4]])
B = np.array([[5,6],
             [7,8]])
print(np.dot(A,B))       # not a element wise multiplication
```
