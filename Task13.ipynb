#**Part C - BroadCasting & Policy Rules**

Policy Rule - if electricity usage on any day is greater than 550 units , a 20% surcharge is applied.

Task C1
create a new array penalized usage such that
- Values > 550 -> multiplied by 1.20
- All other values remain unchanged

Constraints: - Must use np.where. No loops Allowed

````python
# ================================
# Part C: Broadcasting & Policy Rule
# ================================
import numpy as np

# ================================
# Step 1: Create electricity array
# ================================

# Generate 4x7 random electricity usage data
electricity = np.random.randint(120, 651, (4, 7))

print("Electricity Usage:")
print(electricity)




# Policy:
# If electricity usage > 550 â†’ apply 20% surcharge

# Task C1: Create penalized usage array

penalized_usage = np.where(
    electricity > 550,     # condition: check values greater than 550
    electricity * 1.20,    # apply 20% surcharge
    electricity            # keep original values
)

# Display results
print("Original Electricity Usage:")
print(electricity)

print("\nPenalized Electricity Usage:")
print(penalized_usage)
````

## ðŸ§  Problem Understanding

The university applies a **20% surcharge** when electricity usage exceeds **550 units**.

We must create a new array where:

* Values > 550 â†’ multiplied by **1.20**
* Other values remain unchanged
* No loops allowed

---

## ðŸ” Step 1: Understanding `np.where()`

```python
np.where(condition, value_if_true, value_if_false)
```

This function checks each element and chooses a value based on a condition.

Meaning:

* If condition is **True** â†’ use value_if_true
* Else â†’ use value_if_false

---

## ðŸ” Step 2: Apply the Condition

```python
electricity > 550
```

This creates a **True/False matrix** showing where the surcharge applies.

---

## ðŸ” Step 3: Broadcasting

```python
electricity * 1.20
```

NumPy multiplies **every element** by 1.20 automatically.

This is called **broadcasting** â€” applying a single number to an entire array.

---

## âœ… Final Operation

```python
penalized_usage = np.where(
    electricity > 550,
    electricity * 1.20,
    electricity
)
```

This creates a new matrix with surcharge applied only where needed.

---

* Conditional selection using `np.where()`
* Broadcasting for scalar multiplication
* Vectorized operations without loops
