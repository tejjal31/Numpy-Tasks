```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df)*100).round(2)
missing_table = (
    pd.DataFrame({
        "missing_count" : missing_count,
        "missing_percent" : missing_percent
    })
    .sort_values(by = "missing_percent",ascending = False)
)
missing_table[missing_table["missing_count"]>0].head(15)
```

A table like:

Column	Missing Count	Missing %

Sorted from most missing → least missing

# Lot Frontage --- Neighbourhood - wise Median Imputation

```python
# Compute median frontage per neighbourhood
lot_frontage_median = (
    df.groupby("Neighborhood")["Lot Frontage"]
    .median()
)
#Apply neighbourhood-wise median
df["Lot frontage"] = df.apply(
    lambda row: lot_frontage_median[row["Neighborhood"]]
    if pd.isna(row["Lot Frontage"]) else row["Lot Frontage"],
    axis =1
)
```
