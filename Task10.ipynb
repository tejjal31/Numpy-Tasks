````python
import numpy as np

# Create a marks matrix
marks = np.array([
    [80, 75, 90],
    [60, 55, 70]
])

# Create a backup copy of the marks array
backup = marks.copy()

# Modify the original array
marks[0, 0] = 0

# Print both arrays to compare
print("Original marks after modification:")
print(marks)

print("\nBackup array:")
print(backup)
````

## âš ï¸ Discussion: Why is `backup = marks` Dangerous?

If we write:

```python
backup = marks
```

then both variables point to the **same memory location**.

So:

ðŸ‘‰ Changing `marks` will also change `backup`.

This is dangerous because:

* You may accidentally overwrite original data
* Data preprocessing becomes unreliable
* Bugs become hard to detect
