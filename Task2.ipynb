```python
import numpy as np

# -----------------------------------
# Task 2.1: Increase all sales values by 2
# -----------------------------------
arr2 = np.array([
    [10, 15, 8, 12],
    [5, 7, 4, 6],
    [2, 3, 2, 4]
])
print("Updated sales data (+2):")
print(arr2 + 2)


# -----------------------------------
# Task 2.2: Calculate total revenue for each day
# Multiply cost price array by daily sales matrix
# -----------------------------------

# Cost price for each day
cost_price = np.array([20, 30, 15, 25])

# Revenue calculation using NumPy broadcasting
revenue = arr2 * cost_price

print("\nRevenue matrix:")
print(revenue)


# -----------------------------------
# Task 2.3: Scaling â€“ Square root of sales data
# -----------------------------------
print("\nSquare root of sales data:")
print(np.sqrt(arr2))
```
