### ğŸ” What is `np.where()`?

`np.where()` is used to apply **conditions on the whole matrix at once** (no loops needed ğŸš€).

ğŸ“Œ Syntax:


### ğŸ¯ Our rule:

* Marks â‰¥ 50 â†’ **1 (Pass)** âœ…
* Marks < 50 â†’ **0 (Fail)** âŒ

## ğŸ§ª Final Combined Code

```python
import numpy as np

# Create marks matrix
marks = np.array([
    [35, 50, 72],
    [28, 90, 45]
])

# Pass/Fail mapping
status = np.where(marks >= 50, 1, 0)

# Data cleaning
marks[marks < 40] = 0

print("Status Matrix:")
print(status)

print("Cleaned Marks Matrix:")
print(marks)
```
