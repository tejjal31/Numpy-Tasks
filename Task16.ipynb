Part F - Memory Safety
- Create ref backup = electricity
- Create safe backup = electricity.copy()
- Modify one value in the original electricity array.
- Observe which backups change.

Explanation Required - Why is ref_backup risky? when is .copy() required?

````python
import numpy as np

# ================================
# Step 1: Create electricity array
# ================================

# Generate 4x7 random electricity usage data
electricity = np.random.randint(120, 651, (4, 7))

print("Electricity Usage:")
print(electricity)

# ================================
# Part F: Memory Safety
# ================================

# Step 1: Create reference backup (shares memory)
ref_backup = electricity

# Step 2: Create safe backup (independent copy)
safe_backup = electricity.copy()

# Step 3: Modify original array
electricity[0, 0] = 999

# Step 4: Observe changes
print("Modified Original:")
print(electricity)

print("\nReference Backup (changed):")
print(ref_backup)

print("\nSafe Backup (unchanged):")
print(safe_backup)

````

## âš ï¸ Why is `ref_backup` risky?

Using reference backup is risky because:

* It can accidentally overwrite important data
* Bugs become hard to detect
* Original data may be lost during processing

Both variables share the same memory.

---

## âœ… When is `.copy()` required?

`.copy()` is required when:

* You want to preserve original data
* You are performing data cleaning or transformations
* You need a safe backup for experiments

It ensures memory independence.
