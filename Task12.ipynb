## ðŸŽ¯ Final Viva Summary

* NumPy arrays store structured numerical data
* `np.mean`, `np.max`, `np.sum` compute statistics
* `np.argmax` finds position of maximum value
* Axis determines direction of calculation


````python
import numpy as np

# ====================================
# Task A1: Create electricity dataset
# ====================================

# Create a 4x7 array with random electricity usage
# Values range from 120 to 650
# Rows = hostels (h1â€“h4)
# Columns = days (Monâ€“Sun)

electricity = np.random.randint(120, 651, (4, 7))

print("Electricity Usage Array:")
print(electricity)


# ====================================
# Task A2: Print array attributes
# ====================================

print("\nShape:", electricity.shape)
print("Number of dimensions:", electricity.ndim)
print("Total number of elements:", electricity.size)
print("Data type:", electricity.dtype)


# ====================================
# Task B1: Hostel-wise Metrics
# ====================================

# Average electricity usage per hostel (row-wise â†’ axis=1)
avg_hostel = np.mean(electricity, axis=1)

# Maximum usage per hostel (row-wise â†’ axis=1)
max_hostel = np.max(electricity, axis=1)

# Hostel with highest average usage
highest_avg_hostel = np.argmax(avg_hostel)

print("\nAverage usage per hostel:", avg_hostel)
print("Maximum usage per hostel:", max_hostel)
print("Hostel with highest average usage (index):", highest_avg_hostel)


# ====================================
# Task B2: Day-wise Metrics
# ====================================

# Total electricity consumption per day (column-wise â†’ axis=0)
total_per_day = np.sum(electricity, axis=0)

# Day with highest total consumption
highest_day = np.argmax(total_per_day)

print("\nTotal consumption per day:", total_per_day)
print("Day with highest total consumption (index):", highest_day)
````

# âš¡ Task B: Hostel-wise and Day-wise Metrics

The electricity array has shape **(4, 7)**:

* Rows â†’ Hostels
* Columns â†’ Days

Understanding the **axis** is the key concept.

---

## ðŸ  Task B1: Hostel-wise Metrics

To calculate values **per hostel**, we operate **row-wise**.

```python
avg_hostel = np.mean(electricity, axis=1)
max_hostel = np.max(electricity, axis=1)
```

* `axis=1` means operate across columns
* We get one result per row (hostel)

The hostel with the highest average:

```python
highest_avg_hostel = np.argmax(avg_hostel)
```

---

## ðŸ“… Task B2: Day-wise Metrics

To calculate values **per day**, we operate **column-wise**.

```python
total_per_day = np.sum(electricity, axis=0)
```

* `axis=0` means operate down rows
* We get one result per column (day)

The day with highest consumption:

```python
highest_day = np.argmax(total_per_day)
```

---

## â­ Axis Memory Trick

* **axis=0 â†’ go down (columns)**
* **axis=1 â†’ go across (rows)**
