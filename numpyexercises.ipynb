Task C1 : Pass / Fail mapping
- Create a status matrix where:
-- Mark >=50 ->1 (Pass)
-- Mark <=50 ->0 (Fail)

Task C2: Data Cleaning
- To standardize failing grades , replace all the marks below 40 with the value of 0.

# ğŸ¯ NumPy Matrix: Pass/Fail Mapping + Data Cleaning

## ğŸ§  Step 1: Create a Marks Matrix

A **matrix** is a 2D NumPy array (rows + columns) used to store structured data.

```python
import numpy as np

marks = np.array([
    [35, 50, 72],
    [28, 90, 45]
])
```

ğŸ‘‰ This is a **2 Ã— 3 marks matrix**.

---

## âœ… Step 2: Pass / Fail Mapping (Using `np.where()`)

### ğŸ” What is `np.where()`?

`np.where()` is used to apply **conditions on the whole matrix at once** (no loops needed ğŸš€).

ğŸ“Œ Syntax:

```
np.where(condition, value_if_true, value_if_false)
```

ğŸ‘‰ It checks every element and assigns values based on the condition.

### ğŸ¯ Our rule:

* Marks â‰¥ 50 â†’ **1 (Pass)** âœ…
* Marks < 50 â†’ **0 (Fail)** âŒ

```python
status = np.where(marks >= 50, 1, 0)
print(status)
```

ğŸ“Š Output:

```
[[0 1 1]
 [0 1 0]]
```

ğŸ‘‰ Each mark is converted into **Pass/Fail status**.

---

## ğŸ§¹ Step 3: Data Cleaning (Boolean Indexing)

We want to standardize failing grades:

ğŸ‘‰ All marks **below 40 â†’ replace with 0**

```python
marks[marks < 40] = 0
print(marks)
```

ğŸ“Š Output:

```
[[ 0 50 72]
 [ 0 90 45]]
```

ğŸ’¡ This uses **boolean indexing** to update values instantly.

---

## ğŸ§ª Final Combined Code

```python
import numpy as np

# Create marks matrix
marks = np.array([
    [35, 50, 72],
    [28, 90, 45]
])

# Pass/Fail mapping
status = np.where(marks >= 50, 1, 0)

# Data cleaning
marks[marks < 40] = 0

print("Status Matrix:")
print(status)

print("Cleaned Marks Matrix:")
print(marks)
```

---

## ğŸš€ Why Use `np.where()`?

`np.where()` is powerful because it:

âœ¨ Applies conditions to the entire matrix at once
âš¡ Avoids slow loops
ğŸ“Š Is widely used in ML data preprocessing
ğŸ¯ Helps in classification & feature transformation

ğŸ‘‰ In short: **Fast + Clean + Smart coding!**


````python
import numpy as np
# Create a 2D matrix of student marks
marks = np.array([
    [35,50,72],
    [28,90,45]
])
# -----------------------------
# Task C1: Pass / Fail mapping
# -----------------------------

# Create a status matrix:
# Marks >= 50 â†’ 1 (Pass)
# Marks < 50 â†’ 0 (Fail)
status = np.where(marks >= 50, 1, 0)

# Print pass/fail matrix
print("Status Matrix:")
print(status)
````

````python
# -----------------------------
# Task C2: Data Cleaning
# -----------------------------

# Replace all marks below 40 with 0
# This standardizes failing grades
marks[marks < 40] = 0

# Print cleaned marks matrix
print("Cleaned Marks Matrix:")
print(marks)
````
