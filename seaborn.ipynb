**Seaborn** is a Python data visualization library used to make beautiful, statistical graphs easily.

It is built on top of Matplotlib, so it works smoothly with NumPy and Pandas.

```python
import seaborn as sns
```

**distplot** visualizes the distribution of a variable using a histogram and a KDE curve.

```python
sns.distplot(tips['total_bill'])
```

üîπ KDE (Kernel Density Estimation)
üëâ Meaning:

KDE is a smooth line that shows the probability distribution of your data.

1Ô∏è‚É£ Right Skewed (Positive Skew)

2Ô∏è‚É£ Left Skewed (Negative Skew)

```python
sns.distplot(tips['total_bill'],kde = False , bins=30)
```
