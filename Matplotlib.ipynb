```python
import numpy as np
#11 evely spaced values from 0 to 5(both included) step size = (5-0)/(11-1) =0.5
x = np.linspace(0,5,11)
y = x ** 2 #square every elemnet in x
print(x)
print(y)
```

```python
#plots y vs x as a red line('r' means red)
plt.plot(x,y,'r') 
plt.xlabel('X Axis Title Here') #sets the label for x axis
plt.ylabel('Y Axis Title Here') #sets the label for y axis
plt.title('String Title Here') #sets the title of graph
plt.show() #displays the plot on screen
```

Code	Color
'r'	    Red
'b'	    Blue
'g'	    Green
'k'	    Black
'y'	    Yellow

subplot(r,c,i)
ðŸ§  How subplot(r, c, i) Works ?
Part	Meaning
r	    number of rows
c	    number of columns
i	    current plot position

So:

1, 2, 1 â†’ 1 row, 2 columns, first plot

1, 2, 2 â†’ 1 row, 2 columns, second plot

```python
#create a 1x2 grid and select the 1st subplot
plt.subplot(1,2,1)
#plot y vs x with red dashed line
plt.plot(x,y,'r--') 
#select the 2nd subplot in the same 1x2 grid
plt.subplot(1,2,2)
#plot x vs y with green star-marked line
plt.plot(y,x,'g*-')
```

```python
# Create Figure (empty canvas)
fig = plt.figure()
#add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height (range 0 to 1)
#plot on that set of axes
axes.plot(x,y,'b') #plot y vs x in blue color
axes.set_xlabel('Set X Label') #sets label for x axis
axes.set_ylabel('Set y label') #sets label for y axis
axes.set_title('Set Title') #sets the title of the plot
```
